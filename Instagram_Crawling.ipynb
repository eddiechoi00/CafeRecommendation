```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/" + word
    return url
    
```

```python
def select_first(driver):
    #react-root > section > main > article > div:nth-child(3) > div._9AhH0
    first = driver.find_element_by_xpath("//*[@id='react-root']//section//main//article//div[2]//div//div[1]//div[1]//a//div[1]")
    first.click()
    time.sleep(3)
    
#select_first(driver)
```

```python
import re
from bs4 import BeautifulSoup
import unicodedata   # 추가, 2020.08.20

def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    try:
        content = soup.select('div.C4VMK > span')[0].text
        content = unicodedata.normalize('NFC', content)   # 신규 추가. 2020.08.20
    except:
        content = ''
    tags = re.findall(r'#[^\s#,\\]+', content)  
    date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
    try:
        if soup.select('div.Nm9Fw > button')[0].text == '1 like':
            like = 1
        else:
            like = soup.select('div.Nm9Fw > button > span')[0].text
    except:
        like = 0
    try: 
        place = soup.select('div.M30cS')[0].text
        place = unicodedata.normalize('NFC', place)      # 신규 추가. 2020.08.20
    except:
        place = ''
    data = [content, date, like, place, tags]
    return data

#get_content(driver)
```

```python
def move_next(driver):

    right = driver.find_element_by_css_selector('a.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(2)

#move_next(driver)
```
